```python
import pandas as pd
import numpy as np
```

```python
titanic_df = pd.read_csv('titanic.csv')
```

### 정렬, Aggregation 함수, GroupBy 적용

```python
titanic_sorted = titanic_df.sort_values(by=['Name', 'Age'])
titanic_sorted.head()
```

```python
titanic_sorted = titanic_df.sort_values(by=['Name', 'Age'], ascending=False)
titanic_sorted.head()
```

### Aggregation 함수

```python
titanic_df[['Age', 'Fare']].mean()
```

```python
titanic_df[['Age', 'Fare']].count()
```

```python
titanic_groupby = titanic_df.groupby(by=['Pclass'])
print(type(titanic_groupby))
```

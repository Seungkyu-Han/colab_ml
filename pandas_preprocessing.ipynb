```python
import pandas as pd

titanic_df = pd.read_csv('./titanic.csv')
```

### 결손 데이터 처리하기

```python
titanic_df.isna().head()
```

```python
titanic_df.head()
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Age'].isna().sum()
```

```python
titanic_df['Age'].mean()
```

```python
titanic_df['Age']
```

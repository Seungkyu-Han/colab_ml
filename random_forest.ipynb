```python
import numpy as np
import pandas as pd

df = pd.read_csv('test.csv')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

label = df['Activity']
data = df.drop('Activity', axis=1)

encoder = LabelEncoder()

label_encoded = encoder.fit_transform(label)

X_train, X_test, y_train, y_test = train_test_split(
    data, label_encoded, test_size=0.2, random_state=11
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

import pandas as pd
import warnings

rf_clf = RandomForestClassifier(n_estimators=100, random_state=11, max_depth=8)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'max_depth': [16, 24, 32],
    'min_samples_leaf': [1, 6, 12],
    'min_samples_split': [2, 8, 16],
}

rf_clf = RandomForestClassifier(n_estimators=100, random_state=0, n_jobs=-1)
grid_cv = GridSearchCV(rf_clf, param_grid=params, cv=2, n_jobs=-1)
grid_cv.fit(X_train, y_train)

print('최적 하이퍼 파라미터: \n', grid_cv.best_params_)
print('최고 예측 정확도: \n', grid_cv.best_score_)
```

```python
rf_clf1 = RandomForestClassifier(
    n_estimators=100, min_samples_leaf=6, max_depth=16, min_samples_split=2, random_state=0
)

rf_clf1.fit(X_train, y_train)
pred1 = rf_clf1.predict(X_test)
accuracy1 = accuracy_score(y_test, pred1)
print(accuracy1)
```

### GBM(Gradient Boosting Machine)

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("gb", GradientBoostingClassifier(random_state=0))
])

pipeline.fit(X_train, y_train)

gb_pred = pipeline.predict(X_test)
gb_accuracy = accuracy_score(y_test, gb_pred)

print("GBM 정확도:", gb_accuracy)
```
